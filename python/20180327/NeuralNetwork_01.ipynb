```python
import numpy as np
import matplotlib.pyplot as plt
```

## 항등 함수

```python
def identity(x):
    y = x
    return y
```

```python
print(identity(4))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = identity(x)
plt.plot(x, y)
plt.grid()
# plt.ylim(-0.1, 1.1)
plt.show()
```

## 계단 함수 (Step function)

```python
def step_function(x):
    return np.array(x>0, dtype=np.int)
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```

## 시그모이드 함수 (Sigmoid Function)

```python
def sigmoid(x):
    return 1 / (1 +np.exp(-x))
```

```python
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.grid()
plt.show()
```

```python
x = np.linspace(-8, 8, 100)  # -8과 8사이에 100개의 값 생성
sig = sigmoid(x)  # x array 전달
plt.plot(x, sig)  # matplotlib의 plot 기능을 이용하여 (x, y) 좌표 표수
plt.grid()
plt.show()    # plot를 화면에서 볼 수 있도록  show()
```

## tanh 함수

```python
def tanh(x, diff=False):
    if diff:
        return (1+tanh(x))*(1-tanh(x))
    else:
        return np.tanh(x)
```

```python
x = np.linspace(-8, 8, 100)  # -8과 8사이에 100개의 값 생성
sig = tanh(x)  # x array 전달
plt.plot(x, sig)  # matplotlib의 plot 기능을 이용하여 (x, y) 좌표 표수
plt.grid()
plt.show()    # plot를 화면에서 볼 수 있도록  show()
```
